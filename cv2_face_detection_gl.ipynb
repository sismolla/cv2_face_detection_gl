```python
# 127,255 is for the resolution

ret,thershold_binary = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
ret,thershold_inv = cv2.threshold(image,127,255,cv2.THRESH_BINARY_INV)
ret,thershold_trunc = cv2.threshold(image,127,255,cv2.THRESH_TRUNC)
ret,thershold_tozero = cv2.threshold(image,127,255,cv2.THRESH_TOZERO)
ret,thershold_tozero_env = cv2.threshold(image,127,255,cv2.THRESH_TOZERO_INV)
```

```python
name = ['grey_image','thershold_binary','threshold_inv','threshold_trunc','threshold_tozero','thershold_tozero_inv']
images = image,thershold_binary,thershold_inv,thershold_trunc,thershold_tozero,thershold_tozero_env

for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(name[i])
    plt.xticks([])
    plt.yticks([])

plt.show()
#to increase the display area
plt.rcParams['figure.figsize'] = (10,6)
```

```python
face_part = cv2.CascadeClassifier('/content/drive/MyDrive/Colab Notebooks/haarcascade_frontalface_default.xml')
```

```python
face = face_part.detectMultiScale(image,1.1,4)
for (x,y,w,h) in face:
    cv2.rectangle(img1,(x,y),(x+w,y+h),(255,0,0),5)

cv2_imshow(img1)
cv2.waitKey()
```
